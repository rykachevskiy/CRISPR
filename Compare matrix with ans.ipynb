```python
ans_3_full = []
ans_3_split = []
for seq in ans_3:
    ans_3_split.append([x[:31] for x in seq.split("\t")])
    for sp in seq.split("\t"):
        ans_3_full.append(sp[:31])
```

```python
sp_to_num = {}
num_to_sp = {}
ind = 0

for sp in ans_3_full:
    sp_to_num[sp] = ind
    num_to_sp[ind] = sp
    ind += 1
```

```python
ans_pairs = {}

for seq in ans_3_split:
    for i, (x,y) in enumerate(zip(seq[:-1], seq[1:])):
        ans_pairs[(sp_to_num[x], sp_to_num[y])] = 1
```

```python
def closest(sp, sp_to_num=sp_to_num, tr=5):
    dist = 100
    closest = ''
    for ref in sp_to_num.keys():
        d = ed.eval(ref, sp)   
        if d < dist:
            dist = d
            closest = ref

    if dist > tr:
        return -1
    else:
        return sp_to_num[closest]
```

```python
closest('CAAGTGATATCCATCATCGCATCCAGTGCGC', sp_to_num, 29)
```

```python
pairs_data = {}
for pair in pairs:
    key = (closest(pair[0]), closest(pair[1]))
    if key in pairs_data:
        pairs_data[key] += 1
    else:
        pairs_data[key] = 1
    
```
