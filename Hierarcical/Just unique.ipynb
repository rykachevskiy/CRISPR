```python
import editdistance as ed
from tqdm import tqdm
```

```python
len(pairs)
```

```python
def find_closest(sp, ref):
    dist = 100
    ans = -1
    for k in ref.keys():   
        dist_ = ed.eval(sp, k)
        if dist_ < dist:
            ans = ref.get(k)
            dist = dist_

    if dist > 10:
        ans = -1
    return ans
        
```

```python
P = np.zeros((len(pairs), 2))

for i, pair in tqdm(enumerate(pairs[:])):
    P[i][0] = find_closest(pair[0], ref_to_num)
    P[i][1] = find_closest(pair[1], ref_to_num)
```

```python
import networkx as nx
```

```python
   
def to_graph(gr, clusters_num, edges_num):
    G=nx.Graph()
    G.add_nodes_from(np.arange(clusters_num))
    
    edges = []
    curr_weight = 1000
    #while curr_weight > 100:
    for i in range(edges_num):
        lin_max = np.argmax(gr)
        curr_weight = np.max(gr)
        
        x, y = lin_max // gr.shape[0], lin_max % gr.shape[0]
        gr[x][y] = 0
        gr[y][x] = 0

        G.add_edge(x, y)
        edges.append([x, y])
    
    #arr_to_sp(sp_9)
    return edges, G
```

```python
graph = np.zeros((35,35)).astype(int)

for p in tqdm(P.astype(int)):
    graph[p[0]][p[1]] += 1
 
e,G = to_graph(graph, 35, 18)
```
