```python
def update(answer, source , prefix):
    for i, row in source.iterrows():
        names = row[0]
        if type(names) == str:
            #names = [int(x) for x in names.split(',')]
            names = [names]
        else:
            names = [names]

        for name in names:  
            name_str = str(name)
            for single_name in name_str.split(','):
                mask = pd.notna(row[1:]).values
                array = "\t".join(row[1:].values[mask])

                answer[prefix + single_name] = array
```
